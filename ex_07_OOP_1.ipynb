#### Criação

* Criar uma classe chamada Aluno que tenha os atributos: nome e nota
* Fazer o construtor da classe com dois parâmetros. Adicionar no construtor um print para informar que o aluno foi criado com sucesso.
* Adiciona o método string ( __str__ ), para que ao imprimir se veja a sua informação
* Criar um método chamado **qualificacao** que imprima se o aluno passou ou reprovou com base na sua nota

#### Experimentar

* Criar alguns objectos alunos
* Imprimir os objetos alunos criados (o objecto e seus atributos) e o método de qualificação

````python
# definição da classe Aluno - 
# PS: coloquei um pouco de cor usando hexadecimais p/ ficar um pouco mais divertido

class Aluno:
    def __init__(self, nome, nota): # construtor
        self.nome = nome
        self.nota = nota
        print("--" * 30)
        print(f"\033[0;37mAluno criado com sucesso => {self.nome}")
        

    def __str__(self): # método string para mostrar a informação do aluno
        print(f"\033[0;37mNome do aluno(a) {self.nome} - nota final: [{self.nota}]")
        

    def qualificacao(self): # método para mostrar se o aluno foi aprovado ou não
        if self.nota < 0:
            print("\033[1;31;40mERRO! Nota Inválida")
        elif self.nota < 4:
            print(f"\033[0;33;40mAluno(a) {self.nome}", ">> \033[1;31mREPROVADO(A) ")
        elif self.nota < 6:
            print(f"\033[0;33;40mAluno(a) {self.nome}", ">> \033[1;35mRECUPERAÇÃO ")
        elif self.nota <= 10:
            print(f"\033[0;33;40mAluno(a) {self.nome}", ">> \033[1;34mAPROVADO(A) ")
        else:
            print("\033[1;31;40mERRO! Nota Inválida")
            

# criação do aluno1
aluno1 = Aluno("Setembrino Gonçalves", 6)
# Método String para ver os dados do aluno
aluno1.__str__()
# Método qualificação, para saber se o aluno foi aprovado numa escala de 0 a 10
aluno1.qualificacao()

# criação do aluno2
aluno2 = Aluno("Sobervalda de Albuquerque", 3.2)
aluno2.__str__()
aluno2.qualificacao()

# criaçao do aluno3
aluno3 = Aluno("Creuza Josefa", 5.5)
aluno3.__str__()
aluno3.qualificacao()

# criação do aluno4
aluno4 = Aluno("Diocleciano Mendonça", 8.7)
aluno4.__str__()
aluno4.qualificacao()

````

No exercício anterior, ao fazer um print as notas e nomes, vamos obter a nota e nome do aluno. Como pode ser considerado um problema de segurança, vamos encapsular o código para que não se possa aceder diretamente às variáveis e o acesso tenha de ser através de métodos.
* Fazer as variáveis de classe __privadas__. Lembrete: adicionar hífen duplo antes do nome da variável

#### Experimentar

* Comprovar que apenas podemos aceder às variáveis da classe através dos seus métodos get e set (Podemos aceder diretamente ao__nome e__nota? Podemos modifica-los?)

````python
# classe com atributos privados
class Aluno_private:
    '''
    __nome = str("") # atributos privados
    __nota = float(0)
    '''
    
    def __init__(self, nome, nota): # construtor
        self.__nome = nome
        self.__nota = nota
        print("--" * 30)
        #print(f"Aluno criado com sucesso => {self.__nome}")
        

    # Método Getters
    @property
    def nome(self):
        return self.__nome
    @property
    def nota(self):
        return self.__nota
    
    # Métodos Setters
    @nome.setter
    def nome(self,newNome):
        self.__nome = newNome
    @nota.setter
    def nota(self,newNota):
        self.__nota = newNota

# aluno1
aluno1 = Aluno_private("Carmelita Soares", 8.5)

# Chamada dos dados usando o Getter
print(f"Nome do aluno(a): {aluno1.nome}, nota: {aluno1.nota}")

# Mudando a nota do aluno1  usando o Setter
aluno1.nota = 5.3
print(f"Nome do aluno(a): {aluno1.nome}, nota: {aluno1.nota}")

#Mudando a nome do aluno1
aluno1.nome = "Astrogildo Robervaldo"
print(f"Nome do aluno(a): {aluno1.nome}, nota: {aluno1.nota}")

# aluno2
aluno2 = Aluno_private("Oberlaque da Silveira", 3.6)
print(f"Nome do aluno(a): {aluno2.nome}, nota: {aluno2.nota}")

#Mudando o nome e a nota do aluno2
aluno2.nome = "Marinete de Bragança"
aluno2.nota = 9.9
print(f"Nome do aluno(a): {aluno2.nome}, nota: {aluno2.nota}")
````

````python
class Produto:
    # Construtor
    def __init__(self, codigo, nome, preco, tipo, quantidade):
        self.__codigo = codigo
        self.__nome = nome
        self.__preco = preco
        self.__tipo = tipo
        self.__quantidade = quantidade
        
        
    def __str__(self):  # método string para mostrar o produto
        print(f"---------------------------\n"
              f"Produto Criado:\n"
              f"Código:\t[{self.__codigo}],\n"
              f"Nome:\t{self.__nome}\n"
              f"Preço:\t€{self.__preco:.2f}\n"
              f"Tipo:\t{self.__tipo}")
        
        
    def total_compra(self, quantidade):# Método para definir a quantidade do produto
        print(f"---------------------------")
        total = self.__preco * quantidade
        print(f"Produto: [{self.__nome}]\n"
              f"Preço Unitário: [{self.__preco}]\n"
              f"Quantidade: [{quantidade}]\n"
              f"Preço total = €{total:.2f}")
        
        
    # Getters
    @property
    def codigo(self):
        return self.__codigo
    @property
    def nome(self):
        return self.__nome
    @property
    def preco(self):
        return self.__preco
    @property
    def tipo(self):
        return self.__tipo
    @property
    def quantidade(self):
        return self.__quantidade
    
    
    # Setters
    @codigo.setter
    def codigo(self, new_codigo):
        self.__codigo = new_codigo
    @nome.setter
    def nome(self, new_nome):
        self.__nome = new_nome
    @preco.setter
    def preco(self, new_preco):
        self.__preco = new_preco
    @tipo.setter
    def tipo(self, new_tipo):
        self.__tipo = new_tipo
    @quantidade.setter
    def quantidade(self, new_quantidade):
        self.__quantidade = new_quantidade
        

class Pedido(Produto): # criação classe pedido
    lista_total = []
    def __init__(self, lista_produtos = None, lista_preco = None, lista_quantidade = None): # construtor com as listas
        self.lista_produtos = lista_produtos
        self.lista_preco = lista_preco
        self.lista_quantidade = lista_quantidade
        if lista_produtos == None:
            self.lista_produtos = [0]
        if lista_preco == None:
            self.lista_preco = [0]
        if lista_quantidade == None:
            self.lista_quantidade = [0]
            

    def __str__(self): # método para mostrar a lista de produtos e suas respectivas quantidades
        for posição in range (0, len(self.lista_produtos)):
             print(f'Produto: {self.lista_produtos[posição]:.<25}', end="")
             print( f'Quant.: {self.lista_quantidade[posição]:>2} ')
        print('--' * 37)
        

    # método para mostrar o valor total da compra
    def total_pedido(self, lista_produtos, lista_preco, lista_quantidade):
        self.lista_produtos = lista_produtos
        self.lista_preco = lista_preco
        self.lista_quantidade = lista_quantidade
        #self.lista_preco = [float(elem) for elem in self.lista_preco]
        for posição in range(0, len(self.lista_produtos)):
                self.lista_total.append(self.lista_quantidade[posição] * self.lista_preco[posição])
        print(f'Total do Pedido: €{sum(self.lista_total)}')
        

    # método para mostrar o pedido final com todos os dados + o valor final da compra
    def mostrar_pedido(self, lista_produtos, lista_preco, lista_quantidade):
        print("-" * 73)
        print(f'{"PEDIDO FINAL":^73}')
        print("-" * 73)
        self.lista_produtos = lista_produtos
        self.lista_preco = lista_preco
        self.lista_quantidade = lista_quantidade
        for posição in range(0, len(self.lista_produtos)):
            print(f'Produto: {self.lista_produtos[posição]:.<25}', end="")
            print(f'Quant.: {self.lista_quantidade[posição]} >> Preço(unitário): €{self.lista_preco[posição]:>7.2f}')
        print('--' * 37)
        print(f'Total do Pedido: €{sum(self.lista_total)}')
        print('--' * 37)
        
        
# criação de produtos
produto1 = Produto(1001, "Margarita e o Mestre", 25.5, "Livro", 3)
produto2 = Produto(1002, "Maçã de Alcobaça", 2.20, "HortiFruti", 4)
produto3 = Produto(1003, "MacBook Air", 1229, "Informática", 1)
produto4 = Produto(1004, "Travesseiro", 14.45, "Dormitório", 6)
produto5 = Produto(1005, "Casaco de Couro", 115.99, "Vestuário", 2)
produto6 = Produto(1006, "Laranjas", 2.30, "HortiFruti", 3)
produto7 = Produto(1007, "Telemóvel", 200.90, "HortiFruti", 2)

print('Pedido 1\n'
      + ('--' * 30))
# criação do 1° pedido
pedido1 = Pedido([produto1.nome, produto2.nome, produto4.nome],
                 [produto1.preco, produto2.preco, produto4.preco],
                 [produto1.quantidade, produto2.quantidade, produto4.quantidade])

# mostrando os pedidos
pedido1.__str__()
# valor total do pedido

pedido1.total_pedido([produto1.nome, produto2.nome, produto4.nome],
                    [produto1.preco, produto2.preco, produto4.preco],
                    [produto1.quantidade, produto2.quantidade, produto4.quantidade])

# exibindo o pedido completo
pedido1.mostrar_pedido([produto1.nome, produto2.nome, produto4.nome],
                     [produto1.preco, produto2.preco, produto4.preco],
                     [produto1.quantidade, produto2.quantidade, produto4.quantidade])

print("\n")

print('Pedido 2\n'
      + ('--' * 30))

# criação do 2° pedido
pedido2 = Pedido([produto7.nome, produto3.nome, produto5.nome, produto6.nome],
                 [produto7.preco, produto3.preco, produto5.preco, produto6.preco],
                 [produto7.quantidade, produto3.quantidade, produto5.quantidade, produto6.quantidade])

# mostrando os pedidos
pedido2.__str__()

# valor total do pedido
pedido1.total_pedido([produto7.nome, produto3.nome, produto5.nome, produto6.nome],
                    [produto7.preco, produto3.preco, produto5.preco, produto6.preco],
                    [produto7.quantidade, produto3.quantidade, produto5.quantidade, produto6.quantidade])

# exibindo o pedido completo
pedido2.mostrar_pedido([produto7.nome, produto3.nome, produto5.nome, produto6.nome],
                     [produto7.preco, produto3.preco, produto5.preco, produto6.preco],
                     [produto7.quantidade, produto3.quantidade, produto2.quantidade, produto6.quantidade])
````

### O plano cartesiano

Representa um espaço bidimensional (em 2 dimensões), formado por duas linhas perpendiculares, uma horizontal e outra vertical, que se intersetam num ponto. A linha horizontal é chamada eixo das abcissas ou __eixo X__, enquanto a vertical é chamada eixo das ordenadas ou simplesmente o __eixo Y__.
Quanto ao ponto de interseção, é conhecido como o ponto de origem O. __O plano está dividido em 4 quadrantes.__


<img src="./eixos.png"  />



### Pontos e coordenadas

O objetivo é descrever a posição dos **pontos** no plano sob a forma de **coordenadas**, que são formadas associando o valor do eixo X (horizontal) com o valor do eixo Y (vertical).


A representação de um ponto é simples: **P(X,Y)** onde X e Y são a distância horizontal (esquerda ou direita) e vertical (para cima ou para baixo) 
respetivamente, usando como referência o ponto de origem (0,0), mesmo no centro do plano.

<img src="./Puntos.png"  />


### Vetores no plano

Finalmente, um vetor no plano refere-se a um segmento orientado, gerado a partir de dois pontos diferentes.

Para efeitos práticos, é simplesmente uma linha formada a partir de um ponto inicial na direção de outro ponto final, pelo que se entende que um vetor tem comprimento e direção/sentido.

<img src="./Vector.png"  />


Nesta figura, podemos ver dois pontos A e B que poderíamos definir como se segue:
* **A(x1, y1)** => **A(2, 3)**
* **B(x2, y2)** => **B(7, 5)**

E o vetor seria representado como a diferença entre as coordenadas do segundo ponto em relação ao primeiro (o segundo menos o primeiro):
* **AB = (x2-x1, y2-y1)** => **(7-2, 5-3)** => **(5,2)** 

O que significa:

5 à direita e 2 para cima.

````python
from math import sqrt # importando apenas sqrt, sem trazer toda a biblioteca math

class Ponto:#definindo a classe ponto com 2 coordenadas (2-D)
    
    def __init__(self, x=None, y=None):
        self.__x = x
        self.__y = y
        if x == None and y == None: # Na ausência de dados  default é 0
            self.__x = 0
            self.__y = 0
            
            
    @property
    def x(self):
        return self.__x
    @property
    def y(self):
        return self.__y
    
    @x.setter
    def x(self, new_x):
        self.__x = new_x
    @y.setter
    def y(self, new_y):
        self.__y = new_y
        
        
    def __str__(self): # método str para mostrar as coordenadas
        print(f"Ponto nas coordenadas: x={self.__x}, y={self.__y}")
        
        
    def quadrante(self): # método para mostrar o quadrante onde as coordenadas se encontram
        print(f"O ponto [x{self.x},y{self.__y}] está ", end="")
        if self.__x == 0 and self.__y == 0:
            print('na coordenada de origem [X=0, Y=0]')
        elif self.__x > 0 and self.__y > 0:
            print('no Quadrante I')
        elif self.__x < 0 and self.__y > 0:
            print('no Quadrante II')
        elif self.__x < 0 and self.__y < 0:
            print('no Quadrante III')
        else:
            print('no Quadrante IV')
            
            
    def vetor_2(self, ponto1 = None, ponto2 = None):  # método para cálculo do vetor
        self.ponto1 = ponto1
        self.ponto2 = ponto2
        if self.ponto1 == None and self.ponto2 == None:
            self.ponto1 = 0
            self.ponto2 = 0
        vetor_x = self.ponto2.x - self.ponto1.x
        vetor_y = self.ponto2.y - self.ponto1.y
        print("--" * 25)
        print(f"Coordenadas: [x{self.ponto1.x},y{self.ponto1.y}] e [x{self.ponto2.x},y{self.ponto2.y}]")
        #print(f'Resultado do Vetor: {self.ponto2.x - self.ponto1.x}, {self.ponto2.y - self.ponto1.y}')# sem o valor absoluto
        print(f'Resultado do Vetor: [x{abs(vetor_x)},y{abs(vetor_y)}]') # mostrando o valor absoluto

        
    def vetor(self): # método para cálculo do vetor à partir de um segundo ponto fornecido pelo user
        print("--"* 25)
        print(f"Cálculo de Vetor:")
        x2 = int(input("Insira a coordenada x do segundo ponto: "))
        y2 = int(input("Insira a coordenada y do segundo ponto: "))
        print(f'Resultado do Vetor: {x2 - self.__x}, {y2 - self.__y}')
        print("--" * 25)
        
   
class Retangulo(Ponto): # Classe retângulo
    
    def __init__(self, ponto1 = None, ponto2 = None): # construtor Retângulo
        self.ponto1 = ponto1
        self.ponto2 = ponto2
        if self.ponto1 == None and self.ponto2 == None:
            self.ponto1 = 0
            self.ponto2 = 0
            
            
    @property
    def x(self):
        return self.x
    @property
    def y(self):
        return self.y
    
    @y.setter
    def y(self, new_y):
        self.y = new_y
    @x.setter
    def x(self, new_x):
        self.x = new_x
        

    def __str__(self): # mostra a criação do retângulo
        print(f"Retângulo criado com as coordenadas na diagonal:\n"
              f"[x{self.ponto1.x},y{self.ponto1.y}] e [x{self.ponto2.x},y{self.ponto2.y}]")
        print("--" * 15)
        

    def base(self): # cálculo de base
        print("Base do retângulo:",self.ponto2.x - self.ponto1.x)
        print("--" * 15)
        
        
    def altura(self): # cálculo de altura
        print("Altura do retângulo:", self.ponto2.y - self.ponto1.y)
        print("--" * 15)
        
        
    def area(self):  # cálculo de área
        print("Área do retângulo:",(self.ponto2.x - self.ponto1.x) * (self.ponto2.y - self.ponto1.y))
        print("--" * 15)
        
'''
For debug
# criando as coordenadas
ponto1 = Ponto(2,3)
ponto2 = Ponto(7,5)
# Criação de um objeto retângulo
retangulo1 = Retangulo(ponto1, ponto2)
# testantando as funções com um objeto retângulo
retangulo1.__str__()
retangulo1.base()
retangulo1.altura()
retangulo1.area()
'''

class Distancia(Ponto):# classe distância para poder acrescentar 3 pontos.
    def __init__(self, ponto1, ponto2, ponto3):
        self.ponto1 = ponto1
        self.ponto2 = ponto2
        self.ponto3 = ponto3
        if self.ponto1 == None and self.ponto2 == None and self.ponto3 == None:
            self.ponto1 = 0
            self.ponto2 = 0
            self.ponto3 = 0
            

    def distancia_pontos(self, ponto1, ponto2): # método para saber a distância entre 2 pontos
        self.ponto1 = ponto1
        self.ponto2 = ponto2
        if self.ponto1 == None and self.ponto2 == None:
            self.ponto1 = 0
            self.ponto2 = 0
        print('--' * 27)
        distancia_pontos = sqrt(((self.ponto2.x - self.ponto1.x) ** 2) + ((self.ponto2.y - self.ponto1.y) ** 2))
        print(f"A distância entre os pontos: A:[{self.ponto1.x},{self.ponto1.y}] e B:[{self.ponto2.x},{self.ponto2.y}]"
              f" é: {distancia_pontos:.2f}")
        print('--' * 27)
        

    def distancia_centro(self, ponto1, ponto2, ponto3):# Método para saber qual o ponto mais distante da coordenada 00
        self.ponto1 = ponto1
        self.ponto2 = ponto2
        self.ponto3 = ponto3
        if self.ponto1 == None and self.ponto2 == None and self.ponto3 == None:
            self.ponto1 = 0
            self.ponto2 = 0
            self.ponto3 = 0
        distancia1 = sqrt(((self.ponto1.x) ** 2) + ((self.ponto1.y) ** 2))
        distancia2 = sqrt(((self.ponto2.x) ** 2) + ((self.ponto2.y) ** 2))
        distancia3 = sqrt(((self.ponto3.x) ** 2) + ((self.ponto3.y) ** 2))
        print('--' * 40)
        print(f'Dentre os pontos: [{self.ponto1.x},{self.ponto1.y}], [{self.ponto2.x},{self.ponto2.y}], [{self.ponto3.x},{self.ponto3.y}]')
        if abs(distancia1) > abs(distancia2) and abs(distancia1) > abs(distancia3):
            print(f"O ponto mais distante das coordenada [0,0] é o ponto nas coordenadas [{self.ponto1.x},{self.ponto1.y}]")
        elif abs(distancia2) > abs(distancia1) and abs(distancia2) > abs(distancia3):
            print(f"O ponto mais distante das coordenada [0,0] é o ponto nas coordenadas [{self.ponto2.x},{self.ponto2.y}]")
        else:
            print(f"O ponto mais distante das coordenada [0,0] é o ponto nas coordenadas [{self.ponto3.x},{self.ponto3.y}]")
        print('--' * 40)

        
# Experimentação:

# criação dos pontos
pontoA = Ponto(2,3)
pontoB = Ponto(5,5)
pontoC = Ponto(-3, -1)
pontoD = Ponto(0,0)
pontoE = Ponto(-34, -45)

# Indicação dos quadrantes
pontoA.quadrante()
pontoC.quadrante()
pontoD.quadrante()

#Cálculo de Vetores
vetorAB = Retangulo(pontoA, pontoB)
vetorAB.vetor_2(pontoA, pontoB)

vetorBA = Retangulo(pontoB, pontoA)
vetorBA.vetor_2(pontoB, pontoA)

#Distância dos pontos A e B
distanciaAB = Distancia(pontoA, pontoB, 0)
distanciaAB.distancia_pontos(pontoA,pontoB)

#Distância do ponto 00
distancia_00 = Distancia(pontoA, pontoB, pontoC)
distancia_00.distancia_centro(pontoA, pontoB, pontoC)

# Criação do Retângulo
retanguloAB = Retangulo(pontoA, pontoB)

# Métodos com o retângulo
retanguloAB.__str__()
retanguloAB.base()
retanguloAB.altura()
retanguloAB.area()

````

````python
class Veiculo: # Classe veiculo
    
    def __init__(self, cor=None, rodas=None):
        self.__cor = cor
        self.__rodas = rodas

    @property
    def cor(self):
        return self.__cor
    @property
    def rodas(self):
        return self.__rodas

    @cor.setter
    def cor(self, new_cor):
        self.__cor = new_cor
    @rodas.setter
    def rodas(self, new_rodas):
        self.__rodas = new_rodas

    def __str__(self):
        return f"Objeto Veículo criado:\nCor: {self.__cor}\nRodas: {self.__rodas}\n" + ("-=-" * 15)

        
class Carro(Veiculo):  # Classe carro
    def __init__(self, cor = None, rodas = None, velocidade = 0, cilindrada = None):  # construtor
        super().__init__(cor, rodas)
        self.__velocidade = velocidade
        self.__cilindrada = cilindrada
        
    # Getters e setters
    @property
    def velocidade(self):
        return self.__velocidade
    @property
    def cilindrada(self):
        return self.__cilindrada
    
    @velocidade.setter
    def velocidade(self, new_velocidade):
        self.__velocidade = new_velocidade
    @cilindrada.setter
    def cilindrada(self, new_cilindrada):
        self.__cilindrada = new_cilindrada
        
    def __str__(self):  # print do Objeto
        return (f"CARRO\nCor: {self.cor}\nRodas: {self.rodas}\nVelocidade Km/h: {self.__velocidade}\n"
                f"Cilindradas: {self.__cilindrada}\n" + ("-=-" * 15))
    
        
class Bicicleta(Veiculo): # Classe Bicicleta
    
    def __init__(self, tipo=None, cor=None, rodas=None):
        super().__init__(cor, rodas)
        self.__tipo = tipo
        
    @property
    def tipo(self):
        return self.__tipo
    
    @tipo.setter
    def tipo(self, new_tipo):
        self.__tipo = new_tipo
        
    def __str__(self):
        return (f"BICICLETA\nTipo: {self.__tipo}\nCor: {self.cor}\nRodas: {self.rodas}\n" + ("-=-" * 15))


    
class Camiao(Carro): # Classe Camião
    def __init__(self, cor=None, rodas=None, carga=None, velocidade=0, cilindrada=None):
        super().__init__(cor, rodas, velocidade, cilindrada)
        self.__carga = carga
        
    @property
    def carga(self):
        return self.__carga
    
    @carga.setter
    def carga(self, new_carga):
        self.__carga = new_carga
        
        
    def __str__(self):
        return (f"CAMIÃO\nCarregado: {self.__carga}\nVelocidade: {self.velocidade}\nCilindrada: {self.cilindrada}\n"
                f"Cor: {self.cor}\nRodas: {self.rodas}\n" + ("-=-" * 15))

        
class Moto(Bicicleta): # Classe Moto
    
    def __init__(self, tipo=None, cor=None, rodas=None, velocidade = None, cilindrada = None):
        super().__init__(tipo, cor, rodas)
        self.__velocidade = velocidade
        self.__cilindrada = cilindrada
        
        
    @property
    def velocidade(self):
        return self.__velocidade
    @property
    def cilindrada(self):
        return self.__cilindrada

    @velocidade.setter
    def velocidade(self, new_velocidade):
        self.__velocidade = new_velocidade
    @cilindrada.setter
    def cilindrada(self, new_cilindrada):
        self.__cilindrada = new_cilindrada
               
    def __str__(self):
        return (f"MOTO\nTipo: {self.tipo}\nVelocidade: {self.__velocidade}\nCilindrada: {self.__cilindrada}\n"
                f"Cor: {self.cor}\nRodas: {self.rodas}\n" + ("-=-" * 15))
    

# Criação dos Objetos
carro1 = Carro("Cinza", 4, 30, 1.8)
carro2 = Carro('Preto', 4, 0, 1.3)
carro3 = Carro('Branco', 4, 70, 1.6)

bike1 = Bicicleta("Urbana", "Roxo", 2)
bike2 = Bicicleta("Desportiva", "Prata", 2)
bike3 = Bicicleta("Amarela", "Urbana", 2)

camiao1 = Camiao("Azul", 6, True, 0, 5.2, )
camiao2 = Camiao("Vermelho", 8, False,0, 6.5 )
camiao3 = Camiao("Verde") # sem informações disponíveis

moto1 = Moto("Desportiva", "Vermelho", 2, 80, 1.8)
moto2 = Moto("Urbana", "Roxa", 2, 0, 1.2)
moto3 = Moto("Urbana", "Laranja", 2, 60, 1.3)


# Listas de veículos
lista_veiculos_1 = [carro1, carro2, carro3, moto1, camiao1, bike2]
lista_veiculos_2 = [bike1, carro2, bike2, moto2, bike3, camiao3]
lista_veiculos_3 = [carro1, camiao2, bike3, moto3, camiao1, camiao3]
lista_veiculos_4 = [moto2, carro2, camiao3, moto1, bike1, moto3]


class Filtrar():
    
    def __init__(self, lista, veiculo):
        self.lista = lista
        self.veiculo = veiculo
        
    def filtrar(self):
        lista_nova = []
        for item in self.lista:
            if type(item).__name__ == type(self.veiculo).__name__:
                lista_nova.append(item)
        print(f'Lista nova criada com objetos da classe {type(self.veiculo).__name__}')
        print('--' * 23)
        return lista_nova


# Aplicação do filtro
lista_filtrada1 = Filtrar(lista_veiculos_1, carro1)
print("\n".join(str(veiculo) for veiculo in lista_filtrada1.filtrar()))

print("\n")
lista_filtrada2 = Filtrar(lista_veiculos_2, bike2)
print("\n".join(str(veiculo) for veiculo in lista_filtrada2.filtrar()))

print("\n")
lista_filtrada3 = Filtrar(lista_veiculos_3, camiao3)
print("\n".join(str(veiculo) for veiculo in lista_filtrada3.filtrar()))

print("\n")
lista_filtrada4 = Filtrar(lista_veiculos_4, moto1)
print("\n".join(str(veiculo) for veiculo in lista_filtrada4.filtrar()))   
````
