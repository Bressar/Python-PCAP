### Dependência
é um fenômeno que aparece sempre que você vai usar um software que depende de outro software.
Note que a dependência pode incluir (e geralmente inclui) mais de um nível de desenvolvimento 
de software.

## A lista das principais atividades do pip é a seguinte:

pip help operation – mostra uma breve descrição do pip;

pip list – mostra uma lista dos pacotes atualmente instalados;

pip show nome_do_pacote – mostra informações sobre o pacote_nome, incluindo as dependências do pacote;

pip search qualquer_string – pesquisa nos diretórios do PyPI para encontrar pacotes cujos nomes contenham qualquer_string;

pip install nome – instala nome para todo o sistema (espere problemas se você não tiver direitos administrativos);

pip install --user nome – instala nome apenas para você; nenhum outro usuário da plataforma poderá usá-lo;

pip install -U nome – atualiza um pacote previamente instalado;

pip uninstall nome – desinstala um pacote previamente instalado.
